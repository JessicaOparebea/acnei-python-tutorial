### Exercise 1 — Reverse a list without modifying the original

```python
neurotransmitters = ["dopamine", "serotonin", "acetylcholine", "GABA", "glutamate"]
print(neurotransmitters[::-1])  # slicing creates a new list
print(neurotransmitters)        # original unchanged
```

### Exercise 2 — Count unique spike times with a set

```python
spike_times = [0.1, 0.1, 0.3, 0.5, 0.5, 0.9]
unique_spikes = set(spike_times)
print(f"Unique spike times: {unique_spikes}")
print(f"Count: {len(unique_spikes)}")
```

### Exercise 3 — Recording session dictionary

```python
session = {
    "subject_id":       "sub-01",
    "date":             "2026-05-10",
    "brain_region":     "hippocampus",
    "sampling_rate_hz": 30_000,
    "duration_s":       600,
}

for key, val in session.items():
    print(f"  {key:20s}: {val}")
```

### Exercise 4 — Dict comprehension: keep only numerical values

```python
numerical_only = {k: v for k, v in session.items() if isinstance(v, (int, float))}
print(numerical_only)
```

### Exercise 5 — Map channel IDs to regions with zip()

```python
channel_ids = [1, 2, 3, 4, 5]
regions     = ["V1", "V2", "MT", "LIP", "FEF"]

channel_map = dict(zip(channel_ids, regions))
print(channel_map)
```

### Exercise 6 — Spike counts per neuron using enumerate()

```python
spike_trains = [[0.1, 0.3], [0.2, 0.4, 0.6], [0.1]]

for neuron_id, train in enumerate(spike_trains):
    print(f"Neuron {neuron_id}: {len(train)} spike(s)")
```
