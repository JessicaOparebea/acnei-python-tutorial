### Exercise 1 — Even or odd for 1–10

```python
for i in range(1, 11):
    label = "even" if i % 2 == 0 else "odd"
    print(f"{i}: {label}")
```

### Exercise 2 — Detect burst pairs (ISI < 0.05 s)

```python
spike_times = [0.05, 0.1, 0.35, 0.36, 0.9, 1.2]

print("Burst pairs (ISI < 0.05 s):")
for i in range(len(spike_times) - 1):
    isi = spike_times[i+1] - spike_times[i]
    if isi < 0.05:
        print(f"  t={spike_times[i]:.2f}s → t={spike_times[i+1]:.2f}s  (ISI={isi:.3f}s)")
```

### Exercise 3 — Leaky integrator while loop

```python
v          = -70.0
threshold  = -55.0
drive      = 2.0      # mV added per step
leak_coeff = 0.1      # fraction of deviation that leaks back
v_rest     = -70.0
steps      = 0

while v < threshold and steps < 100:
    v += drive - leak_coeff * (v - v_rest)
    steps += 1

print(f"Final potential: {v:.2f} mV")
print(f"Steps taken: {steps}")
print("Reached threshold!" if v >= threshold else "Did not reach threshold in 100 steps.")
```

### Exercise 4 — Mean firing rate of active neurons only

```python
rates = [0, 5, 0, 12, 0, 8, 0, 3]
total = 0
count = 0

for r in rates:
    if r == 0:
        continue
    total += r
    count += 1

mean_rate = total / count if count > 0 else 0
print(f"Active neurons: {count}")
print(f"Mean firing rate (active only): {mean_rate:.2f} Hz")
```

### Exercise 5 — Multiplication table as a dict of dicts (1–5)

```python
mult_table = {}
for i in range(1, 6):
    mult_table[i] = {}
    for j in range(1, 6):
        mult_table[i][j] = i * j

# Pretty-print
print("   ", "  ".join(str(j) for j in range(1, 6)))
for i in range(1, 6):
    row = "  ".join(f"{mult_table[i][j]:2d}" for j in range(1, 6))
    print(f"{i}:  {row}")
```
