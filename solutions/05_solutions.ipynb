### Exercise 1 — ISI function

```python
def isi(t1, t2):
    """Return the inter-spike interval between two spike times."""
    return abs(t2 - t1)

print(isi(0.1, 0.3))
print(isi(0.5, 0.9))
print(isi(1.0, 1.05))
```

### Exercise 2 — Mean firing rate

```python
def mean_firing_rate(spike_times):
    """
    Compute mean firing rate in Hz from a list of spike times (seconds).
    Returns None if fewer than 2 spikes.
    """
    if len(spike_times) < 2:
        print("Need at least 2 spikes to estimate firing rate.")
        return None
    isis = [spike_times[i+1] - spike_times[i] for i in range(len(spike_times) - 1)]
    mean_isi = sum(isis) / len(isis)
    return 1.0 / mean_isi

print(f"{mean_firing_rate([0.1, 0.2, 0.35, 0.5, 0.65]):.2f} Hz")
print(mean_firing_rate([0.5]))  # edge case
```

### Exercise 3 — Filter spikes within a time window

```python
def filter_spikes(spike_times, t_start, t_end):
    """Return spike times within [t_start, t_end]."""
    return [t for t in spike_times if t_start <= t <= t_end]

spikes = [0.05, 0.1, 0.4, 0.6, 0.85, 1.1, 1.3]
print(filter_spikes(spikes, 0.3, 0.9))
```

### Exercise 4 — describe() with *args

```python
def describe(*values):
    """Print min, max, and mean of any number of numerical values."""
    if not values:
        print("No values provided.")
        return
    print(f"n={len(values)}  min={min(values)}  max={max(values)}  mean={sum(values)/len(values):.3f}")

describe(5, 12, 8, 3, 17, 6)
```

### Exercise 5 — Sort list of tuples by second element using lambda

```python
data = [("theta", 7), ("gamma", 60), ("delta", 2), ("alpha", 10)]
sorted_data = sorted(data, key=lambda item: item[1])
print(sorted_data)
```

### Exercise 6 — apply_threshold() with default argument

```python
def apply_threshold(voltages, threshold=-55):
    """Return a list of booleans: True where voltage exceeds threshold."""
    return [v > threshold for v in voltages]

print(apply_threshold([-80, -65, -55, -40, 30]))
print(apply_threshold([-80, -65, -55, -40, 30], threshold=-65))
```
