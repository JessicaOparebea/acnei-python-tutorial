### Exercise 1 — Action potential amplitude

```python
action_potential_peak = 40    # mV
resting_potential     = -70   # mV
amplitude = action_potential_peak - resting_potential
print(f"Action potential amplitude: {amplitude} mV")
```

### Exercise 2 — String methods on neurotransmitter name

```python
neurotransmitter = "Gamma-Aminobutyric Acid"
print(neurotransmitter.lower())
print(neurotransmitter.upper())
print(neurotransmitter.startswith("gamma"))  # False
# Explanation: .startswith() is case-sensitive.
# The string begins with uppercase 'G', not lowercase 'g'.
# neurotransmitter.lower().startswith("gamma") would return True.
```

### Exercise 3 — type(True + True)

```python
print(type(True + True))  # <class 'int'>
print(True + True)        # 2
# Explanation: In Python, bool is a subclass of int.
# True == 1 and False == 0, so True + True == 1 + 1 == 2.
```

### Exercise 4 — Convert string to int and print with comma formatting

```python
print(f"{int('86000000000'):,}")
```

### Exercise 5 — Catch ValueError from int("hello")

```python
try:
    int("hello")
except ValueError as e:
    print(f"Cannot convert 'hello' to an integer: {e}")
    print("Tip: int() only accepts strings that contain digits, e.g. int('42').")
```
