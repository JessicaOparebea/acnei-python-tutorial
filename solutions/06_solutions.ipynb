```python
import numpy as np
import matplotlib.pyplot as plt
```

### Exercise 1 — 5 Hz sine wave + Gaussian noise, both plotted on the same axes

```python
fs    = 1000                              # samples/second
t     = np.linspace(0, 2, fs * 2)        # 2-second time axis
clean = np.sin(2 * np.pi * 5 * t)        # 5 Hz sine
noisy = clean + np.random.normal(0, 0.3, size=len(t))

fig, ax = plt.subplots(figsize=(11, 3))
ax.plot(t, noisy,  color='lightcoral',  linewidth=0.8, label='Noisy signal', alpha=0.8)
ax.plot(t, clean,  color='royalblue',   linewidth=1.8, label='Clean 5 Hz')
ax.set_title("5 Hz Sine Wave with Gaussian Noise")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Exercise 2 — PSTH from 20 simulated neurons, 20 ms bins

```python
np.random.seed(0)
n_neurons  = 20
all_spikes = np.concatenate([
    np.random.uniform(0, 1, np.random.randint(5, 25))
    for _ in range(n_neurons)
])

bin_edges   = np.arange(0, 1.02, 0.02)      # 20 ms bins
counts, edges = np.histogram(all_spikes, bins=bin_edges)
bin_centres = (edges[:-1] + edges[1:]) / 2

fig, ax = plt.subplots(figsize=(10, 3))
ax.bar(bin_centres, counts, width=0.018, color='steelblue', edgecolor='white')
ax.set_title("PSTH — 20 Neurons, 20 ms Bins")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Spike count")
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Exercise 3 — 8-channel recording: mean ± std with fill_between()

```python
np.random.seed(1)
recording = np.random.randn(8, 1000)
time_axis = np.arange(1000)
chan_mean  = recording.mean(axis=0)
chan_std   = recording.std(axis=0)

fig, ax = plt.subplots(figsize=(11, 3))
ax.plot(time_axis, chan_mean, color='royalblue', linewidth=1.5, label='Mean across channels')
ax.fill_between(time_axis,
                chan_mean - chan_std,
                chan_mean + chan_std,
                alpha=0.25, color='royalblue', label='±1 SD')
ax.set_title("8-Channel Recording: Mean ± SD")
ax.set_xlabel("Sample")
ax.set_ylabel("Amplitude")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Exercise 4 — Power spectrum of 10 Hz alpha oscillation

```python
fs      = 500
t       = np.linspace(0, 1, fs, endpoint=False)
signal  = np.sin(2 * np.pi * 10 * t)

spectrum       = np.abs(np.fft.rfft(signal))
freqs          = np.fft.rfftfreq(len(t), d=1/fs)
dominant_freq  = freqs[np.argmax(spectrum)]

fig, ax = plt.subplots(figsize=(9, 3))
ax.plot(freqs, spectrum, color='darkorange', linewidth=1.5)
ax.axvline(dominant_freq, color='royalblue', linestyle='--', linewidth=1.2,
           label=f'Peak: {dominant_freq:.0f} Hz')
ax.set_title("Power Spectrum — 10 Hz Alpha Oscillation")
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("Amplitude")
ax.set_xlim(0, 50)
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
