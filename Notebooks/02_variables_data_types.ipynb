## 2.3 Arithmetic

```python
# Standard operators
print(10 + 3)   # addition
print(10 - 3)   # subtraction
print(10 * 3)   # multiplication
print(10 / 3)   # division (always returns float)
print(10 // 3)  # floor division (integer result)
print(10 % 3)   # modulo (remainder)
print(10 ** 3)  # exponentiation
```

/ vs //   —  Division always returns a float: 10 / 2 = 5.0, not 5. Floor division discards the decimal: 10 // 3 = 3.

% (modulo)   —  Returns what’s left over after dividing. 10 % 3 = 1 because 10 = 3×3 + 1. Useful for cycling (every Nth trial) and checking even/odd.

** (exponentiation)   —  Raises to a power. 10 ** 3 = 1000. Comes up in normalisations and signal scaling.

```python
# Augmented assignment operators
count = 0
count += 1   # same as count = count + 1
count += 1
count *= 10  # same as count = count * 10
print(count)
```
