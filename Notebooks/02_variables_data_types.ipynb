## 2.1 Basic Types

Python has a small set of built-in types for representing data.

```python
# Integers — whole numbers
type(86_000_000_000)
```

```python
# Floats — decimal numbers
type(-70.0)
```

```python
# Adding int + float gives float
type(1 + 1.0)
```

```python
# Floating point is an approximation!
0.1 + 0.2
```

> **Note:** `0.1 + 0.2` does not equal `0.3` exactly — this is a fundamental limitation of how computers store decimal numbers (floating point representation). For most neuroscience work this won't matter, but be aware when doing equality comparisons.

```python
# Strings — text
type("hippocampus")
```

```python
# Both quote styles are equivalent
"hippocampus" == 'hippocampus'
```

```python
# Booleans — True or False
type(True), type(False)
```

```python
# None — represents the absence of a value
type(None)
```

## 2.2 Variables

A variable is a name that points to a value. Assignment uses `=`.

```python
resting_potential   = -70.0    # mV
threshold_potential = -55.0    # mV
brain_region        = "cortex"
is_excitatory       = True

print(resting_potential)
print(brain_region)
```

```python
# Variable names: lowercase_with_underscores is the Python convention
# Good:
spike_rate = 40
neuron_count = 100

# Avoid:
# SpikeRate = 40     (that's a class name convention)
# sr = 40            (too cryptic)
```

```python
# You can reassign a variable
x = 10
print(x)
x = 99
print(x)
```

```python
# Multiple assignment on one line
a, b, c = 1, 2, 3
print(a, b, c)
```

## 2.3 Arithmetic

```python
# Standard operators
print(10 + 3)   # addition
print(10 - 3)   # subtraction
print(10 * 3)   # multiplication
print(10 / 3)   # division (always returns float)
print(10 // 3)  # floor division (integer result)
print(10 % 3)   # modulo (remainder)
print(10 ** 3)  # exponentiation
```

```python
# Practical example: voltage change to threshold
voltage_change = threshold_potential - resting_potential
print(f"A neuron needs {voltage_change} mV to reach threshold")
```

```python
# Augmented assignment operators
count = 0
count += 1   # same as count = count + 1
count += 1
count *= 10  # same as count = count * 10
print(count)
```

## 2.4 Strings

```python
region = "hippocampus"

# Useful string methods
print(region.upper())
print(region.capitalize())
print(region.replace("hippocampus", "amygdala"))
print(region.startswith("hippo"))
print(len(region))
```

```python
# In Jupyter, type a dot after a string and press TAB to see all methods
# region.<TAB>
```

```python
# f-strings: embed variables or expressions directly in text
potential = -70.0
region    = "cortex"

print(f"The {region} neuron has a resting potential of {potential} mV")
print(f"Threshold gap: {-55 - potential} mV")
print(f"{potential = }")  # handy debug format: shows name AND value
```

```python
# Format numbers inside f-strings
pi_approx = 3.14159265
print(f"{pi_approx:.2f}")   # 2 decimal places
print(f"{pi_approx:.4f}")   # 4 decimal places
print(f"{1_000_000:,}")     # thousands separator
```

## 2.5 Type Conversion

```python
# Converting between types
print(int(3.9))    # → 3 (truncates, does NOT round)
print(float(5))    # → 5.0
print(str(42))     # → '42'
print(bool(0))     # → False
print(bool(1))     # → True
print(bool(""))    # → False (empty string is falsy)
print(bool("hi"))  # → True
```

```python
# Common gotcha: adding a number and a string
try:
    print(1 + "1")
except TypeError as e:
    print("TypeError:", e)
    print("You need to convert types explicitly:")
    print(1 + int("1"))
```
