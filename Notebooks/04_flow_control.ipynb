## 4.1 if / elif / else

Conditional logic: run a block of code only when a condition is true.

```python
#If and Else Statement
potential_mv = -60

if potential_mv >= -55:
    print("Neuron fires!")
else:
    print("Below threshold — no action potential")
```

```python
# Multiple conditions with elif
for voltage in [-80, -60, -40, 20]:

    if voltage < -70:
        print(f"{voltage} mV → resting")

    elif voltage < -50:
        print(f"{voltage} mV → weak signal")

    else:
        print(f"{voltage} mV → neuron firing")
```

```python
# Comparison operators
print(5 == 5)   # equal
print(5 != 4)   # not equal
print(5 > 3)    # greater than
print(5 >= 5)   # greater than or equal
print(3 < 5 < 10)   # chained comparison!
```

```python
# Logical operators: and, or, not
is_excitatory = True
is_firing     = True

print(is_excitatory and is_firing)    # both must be True
print(is_excitatory or False)         # at least one must be True
print(not is_firing)                  # invert
```

```python
# Inline conditional (ternary)
v = -60
state = "firing" if v >= -55 else "silent"
print(state)
```

## 4.2 for loops

Iterate over any sequence: list, tuple, string, dictionary, range...

```python
spike_times = [0.1, 0.3, 0.5, 0.9]

for t in spike_times:
    print(f"Spike at t = {t:.1f} s")
```

```python
# range() generates a sequence of numbers
for i in range(5):
    print(i, end=" ")
print()

for i in range(2, 10, 2):   # start, stop, step
    print(i, end=" ")
print()
```

```python
# enumerate() — index + value
regions = ["cortex", "hippocampus", "amygdala"]

for i, region in enumerate(regions):
    print(f"  Electrode {i+1}: {region}")
```

```python
# zip() — iterate over two sequences in parallel
channels = ["ch1", "ch2", "ch3"]
rates    = [32, 45, 18]

for channel, rate in zip(channels, rates):
    print(f"{channel}: {rate} Hz")
```

```python
# Accumulating results in a loop
spike_times = [0.1, 0.3, 0.35, 0.7, 0.9]
isis = []

for i in range(len(spike_times) - 1):
    isi = spike_times[i+1] - spike_times[i]
    isis.append(round(isi, 3))

print("ISIs:", isis)
```

## 4.3 while loops

Keep running as long as a condition is True.

```python
# Simulate membrane potential charging toward threshold
potential   = -70.0
threshold   = -55.0
input_current = 1.5    # mV per step
steps = 0

while potential < threshold:
    potential += input_current
    steps += 1

print(f"Reached threshold in {steps} steps")
print(f"Final potential: {potential:.1f} mV")
```

> **Note:** Always make sure the condition in a `while` loop will eventually become `False`. An infinite loop will freeze your kernel — you can interrupt it with *Kernel → Interrupt*.

## 4.4 break, continue, pass

```python
# break — exit the loop immediately
voltages = [-70, -65, -58, -52, -40, 30]

for v in voltages:
    if v > -55:
        print(f"First suprathreshold value: {v} mV")
        break
```

```python
# continue — skip the rest of this iteration
for v in voltages:
    if v >= -55:     # skip firing values
        continue
    print(f"Subthreshold: {v} mV")
```
