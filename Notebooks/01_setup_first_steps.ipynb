## 1.1 What is Python?

Python is a general purpose programming language that has become the standard tool in data science, neuroscience and AI. It is:

- Free and open source
- Easy to read and write
- Backed by a huge ecosystem of scientific libraries

In neuroscience, Python is used for everything from analysing spike trains to building neural network simulations.

> **Note:** Python was named after Monty Python, the British comedy group, not the snake. Although the snake logo stuck.

## 1.2 What is a Jupyter Notebook?

A Jupyter Notebook is an interactive environment that mixes:
- **Code cells** — Python code you can run
- **Markdown cells** — formatted text, like this one

**To run a cell:** click it, then press `Shift + Enter`

## 1.5 print()

`print()` is a built-in function that displays output.

```python
print("Hello, neuroscience world!")
```

```python
# You can print multiple things at once
print("The brain has roughly", 86_000_000_000, "neurons")
```

```python
# Print with a custom separator
print("cortex", "hippocampus", "cerebellum", sep=" | ")
```

```python
# In a notebook, the last expression in a cell is shown automatically
# (no print needed)
1 + 1
```

```python
# But if you want to show multiple things, you need print()
print(1 + 1)
print(2 + 2)
3 + 3  # only this last one shows automatically
```

> **Tip:** Get in the habit of using `print()` it works the same inside scripts, notebooks and anywhere else.
