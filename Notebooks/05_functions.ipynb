## 5.1 Basic Syntax

```python
def will_fire(membrane_potential, threshold):
    """Return True if the neuron reaches threshold."""
    return membrane_potential >= threshold

will_fire(-60, -55)
```

```python
will_fire(-40, -55)
```

> **Important:** Python functions **always** return a value. Without an explicit `return`, they return `None`.

## 5.2 Return Values

```python
# Return multiple values — Python packs them into a tuple
def spike_stats(spike_times):
    """
    Compute basic statistics from a list of spike times.
    Returns (n_spikes, mean_isi, firing_rate_hz).
    """
    n = len(spike_times)
    if n < 2:
        return n, None, None
    isis = [spike_times[i+1] - spike_times[i] for i in range(n-1)]
    mean_isi = sum(isis) / len(isis)
    rate_hz  = 1.0 / mean_isi
    return n, mean_isi, rate_hz

n, isi, rate = spike_stats([0.1, 0.2, 0.35, 0.5, 0.65, 0.8])
print(f"n={n}  mean_ISI={isi:.3f}s  rate={rate:.1f} Hz")
```

```python
# Early return — exits as soon as it hits a return
def classify_potential(v):
    if v < -70:
        return "hyperpolarised"
    if v < -55:
        return "subthreshold"
    if v < 0:
        return "depolarising"
    return "peak"

classify_potential(-77)
```

## 5.3 Default Arguments

```python
# threshold has a default value — it's optional when calling
def will_fire(membrane_potential, threshold=-55.0):
    """Return True if the neuron reaches threshold."""
    return membrane_potential >= threshold

will_fire(-50)              # uses default threshold
```

```python
will_fire(-50, threshold=-48)   # override default
```

> **Tip:** Use keyword arguments when calling functions — it makes code much more readable:
>
> `will_fire(-50, threshold=-48)` is clearer than `will_fire(-50, -48)`

```python
# WARNING: Never use mutable defaults (e.g. lists) as default arguments!

# This is a classic Python bug:
def bad_append(item, container=[]):    # DON'T do this
    container.append(item)
    return container

print(bad_append(1))  # [1]
print(bad_append(2))  # [1, 2] — NOT [2]!

# Correct pattern:
def good_append(item, container=None):
    if container is None:
        container = []
    container.append(item)
    return container

print(good_append(1))  # [1]
print(good_append(2))  # [2]
```

```python
lis = [2]
lis.append(3)
print(lis)
```

## 5.4 *args and **kwargs

```python
# *args: accept any number of positional arguments → packed as a tuple
def total_spikes(*spike_counts):
    print(spike_counts)
    return sum(spike_counts)

total_spikes(5, 12, 8)
```

```python
total_spikes(3, 7, 2, 14, 9)
```

```python
# **kwargs: accept any keyword arguments → packed as a dict
def describe_neuron(**properties):
    for key, val in properties.items():
        print(f"  {key}: {val}")

describe_neuron(region="cortex", type="pyramidal", Vrest=-70)
```

```python
# Combining both
def general(*args, **kwargs):
    print("args:", args)
    print("kwargs:", kwargs)

general(1, 2, region="cortex", rate=40)
```

## 5.5 Functions are Values

In Python, functions are first-class objects — you can pass them as arguments.

```python
def apply_to_each(data, func):
    """Apply func to every element of data."""
    return [func(x) for x in data]

voltages = [-70, -65, -55, -40, 30]

# Pass built-in abs
print(apply_to_each(voltages, abs))

# Pass a custom function
def mv_to_v(mv):
    return mv / 1000

print(apply_to_each(voltages, mv_to_v))
```

## 5.6 Anonymous Functions (lambda)

```python
# Syntax: lambda [args]: [expression]
mv_to_v = lambda mv: mv / 1000
mv_to_v(-70)
```

```python
# Most useful inline, as an argument to another function
print(apply_to_each(voltages, lambda mv: mv / 1000))
```

```python
# sorted() with a custom key - takes in iterables and perform a sort
neurons = [
    {"id": 1, "rate_hz": 32},
    {"id": 2, "rate_hz": 8},
    {"id": 3, "rate_hz": 47},
]

# Sort by firing rate
sorted_neurons = sorted(neurons, key=lambda n: n["rate_hz"])
for n in sorted_neurons:
    print(f"Neuron {n['id']}: {n['rate_hz']} Hz")
```

> **Tip:** If your lambda is longer than one line or you find yourself reusing it, define a proper `def` function instead.
