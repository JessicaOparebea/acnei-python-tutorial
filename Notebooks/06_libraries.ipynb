## 6.2 math

```python
import math

print(math.pi)
print(math.e)
print(math.sqrt(2))
print(math.log(100, 10))     # log base 10
print(math.sin(math.pi / 2)) # sin(90°) = 1.0
```

```python
# Exponential decay — relevant for membrane time constants
tau = 20.0    # membrane time constant (ms)
t   = 10.0    # time since stimulus (ms)

decay = math.exp(-t / tau)
print(f"After {t} ms (τ={tau} ms), fraction remaining: {decay:.4f}")
```

## 6.3 NumPy

NumPy provides fast multi-dimensional arrays and mathematical operations on them. It is the foundation of scientific Python.

```python
import numpy as np

# Creating arrays
v = np.array([-70, -65, -55, -40, 30, -60, -70])  # from list
t = np.linspace(0, 1, 100)                          # 100 evenly-spaced points
z = np.zeros(5)                                     # [0. 0. 0. 0. 0.]

print(v)
print(f"shape: {v.shape}, dtype: {v.dtype}")
```

```python
# Element-wise math — no loop needed
v_in_volts = v / 1000
print(v_in_volts)

# Works with any operation
print(v + 10)
print(v ** 2)
```

```python
# Boolean indexing — very powerful
above_threshold = v[v > -55]
print("Above threshold:", above_threshold)
```

```python
# Statistics
data = np.array([12.3, 15.1, 10.9, 18.4, 14.0, 11.8, 16.5, 13.2])

print(f"mean:   {np.mean(data):.2f}")
print(f"std:    {np.std(data):.2f}")
print(f"min:    {np.min(data):.2f}")
print(f"max:    {np.max(data):.2f}")
print(f"median: {np.median(data):.2f}")
```

```python
# 2D arrays — useful for multi-channel recordings
# Rows = channels, Columns = time points
recording = np.random.randn(4, 100)   # 4 channels, 100 time points
print("Shape:", recording.shape)
print("Channel 0, first 5 values:", recording[0, :5].round(3))
print("Mean per channel:", recording.mean(axis=1).round(3))
```

```python
# Generating a sine wave — e.g. an EEG oscillation
t   = np.linspace(0, 1, 500)    # 1 second
f   = 10                         # 10 Hz (alpha)
lfp = np.sin(2 * np.pi * f * t)

print("LFP shape:", lfp.shape)
print("Max:", lfp.max(), "Min:", lfp.min())
```

## 6.4 Matplotlib

```python
import matplotlib.pyplot as plt

# Basic line plot
t   = np.linspace(0, 1, 500)
lfp = np.sin(2 * np.pi * 10 * t)

plt.figure(figsize=(10, 3))
plt.plot(t, lfp, color='royalblue', linewidth=1.5)
plt.title("10 Hz Alpha Oscillation")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Raster plot — a classic neuroscience figure
np.random.seed(42)
n_neurons = 10

fig, ax = plt.subplots(figsize=(10, 4))
for nid in range(n_neurons):
    n_spikes    = np.random.randint(5, 20)
    spike_times = np.sort(np.random.uniform(0, 1, n_spikes))
    ax.scatter(spike_times, [nid] * n_spikes,
               marker='|', s=200, color='black', linewidths=1.2)

ax.set_title("Raster Plot — 10 Neurons")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Neuron ID")
ax.set_yticks(range(n_neurons))
ax.grid(True, axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Subplots — show multiple panels side by side
t = np.linspace(0, 1, 500)

fig, axes = plt.subplots(1, 3, figsize=(14, 3))

for ax, freq, name in zip(axes, [4, 10, 40], ["Theta (4 Hz)", "Alpha (10 Hz)", "Gamma (40 Hz)"]):
    signal = np.sin(2 * np.pi * freq * t)
    ax.plot(t[:100], signal[:100])
    ax.set_title(name)
    ax.set_xlabel("Time (s)")
    ax.grid(True, alpha=0.3)

axes[0].set_ylabel("Amplitude")
plt.suptitle("EEG Frequency Bands", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Histogram
np.random.seed(7)
isis = np.random.exponential(scale=0.05, size=500)

plt.figure(figsize=(7, 3))
plt.hist(isis, bins=40, color='steelblue', edgecolor='white')
plt.title("ISI Distribution (Simulated Poisson Neuron)")
plt.xlabel("ISI (s)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
