We often want to store and work with *collections* of data — a list of spike times, a mapping from neuron ID to brain region, a set of unique conditions. Python provides four built-in data structures for this.

| Structure | Syntax | Ordered | Mutable | Unique items |
|-----------|--------|---------|---------|---------------|
| `list` | `[1, 2, 3]` | ✅ | ✅ | ❌ |
| `tuple` | `(1, 2, 3)` | ✅ | ❌ | ❌ |
| `dict` | `{"k": v}` | ✅ (Python 3.7+) | ✅ | ❌ (keys only) |
| `set` | `{1, 2, 3}` | ❌ | ✅ | ✅ |

## 3.1 Lists

A `list` is an ordered, mutable sequence. It is the most commonly used data structure in Python.

```python
# Creating lists
brain_regions = ["cortex", "hippocampus", "cerebellum", "amygdala", "thalamus"]
spike_times   = [0.1, 0.3, 0.5, 0.9, 1.2, 1.8]  # seconds
mixed         = ["cortex", 40, True, None]        # lists can hold any types

print(brain_regions)
print(f"Number of regions: {len(brain_regions)}")
```

```python
# Indexing — starts at 0!
print(brain_regions[0])    # first
print(brain_regions[2])    # third
print(brain_regions[-1])   # last
print(brain_regions[-2])   # second to last
```

```python
# Slicing — list[start:stop:step] (stop is EXCLUDED)
print(brain_regions[1:3])   # items at index 1 and 2
print(brain_regions[:2])    # first two
print(brain_regions[2:])    # from index 2 to end
print(brain_regions[::2])   # every second item
print(brain_regions[::-1])  # reversed
```

```python
# Lists are MUTABLE — you can change them in place
brain_regions[0] = "prefrontal cortex"
print(brain_regions)
```

```python
# Useful list methods
regions = ["cortex", "hippocampus", "amygdala"]

regions.append("thalamus")       # add to end
print("append:", regions)

regions.insert(1, "cerebellum")  # insert at position
print("insert:", regions)

regions.remove("amygdala")       # remove first occurrence of value
print("remove:", regions)

last = regions.pop()              # remove and return last item
print("popped:", last)
print("after pop:", regions)

regions.sort()
print("sorted:", regions)
```

```python
# Checking membership
print("hippocampus" in regions)
print("thalamus" in regions)
```

```python
# Looping
for region in regions:
    print(region)

print()

# With enumerate() to get index + value
for i, region in enumerate(regions):
    print(f"{i}: {region}")
```

```python
# List comprehension — a compact way to build lists
voltages = [-80, -70, -65, -55, -40, 30]

# All voltages above threshold
above_threshold = [v for v in voltages if v > -55]
print(above_threshold)

# Convert mV to V
in_volts = [v / 1000 for v in voltages]
print(in_volts)
```

> **Important — Lists hold references, not copies:**

```python
a = [1, 2, 3]
b = a           # b points to the SAME list
b[0] = 99
print(a)        # [99, 2, 3] — a was also changed!

# To make a copy:
c = a.copy()
c[0] = 0
print(a)        # [99, 2, 3] — a is unchanged now
print(c)        # [0, 2, 3]
```

## 3.2 Tuples

Tuples are like lists but **immutable** — they cannot be modified after creation. Use them for data that should stay fixed.

```python
electrode_xyz = (34.5, -12.3, 8.0)   # x, y, z in mm — shouldn't change
print(electrode_xyz)
print(electrode_xyz[0])  # indexing works the same way
```

```python
# Tuples cannot be modified
try:
    electrode_xyz[0] = 99
except TypeError as e:
    print("TypeError:", e)
```

```python
# Tuple unpacking — very clean and Pythonic
x, y, z = electrode_xyz
print(f"Electrode: x={x}, y={y}, z={z}")

# Unpacking with *rest
first, *rest = (1, 2, 3, 4, 5)
print(first, rest)
```

```python
# Functions often return multiple values as tuples
def min_max(values):
    return min(values), max(values)

low, high = min_max([10, 3, 7, 1, 8])
print(f"Range: {low} to {high}")
```

## 3.3 Dictionaries

A dictionary maps **keys** to **values**. Keys must be unique and immutable (strings, numbers, tuples). Values can be anything.

```python
neuron = {
    "type":      "pyramidal",
    "region":    "cortex",
    "Vrest_mV":  -70.0,
    "Vthr_mV":   -55.0,
    "excitatory": True,
}

neuron
```

```python
# Accessing values by key
print(neuron["type"])
print(neuron["Vrest_mV"])
```

```python
# Accessing a missing key raises a KeyError
try:
    print(neuron["axon_length"])
except KeyError as e:
    print("KeyError:", e)

# .get() is safer — returns None (or a default) if key is missing
print(neuron.get("axon_length", "unknown"))
```

```python
# Add / update keys
neuron["firing_rate_hz"] = 40
neuron["Vrest_mV"]       = -72   # update
print(neuron)
```

```python
# Iterating
for key in neuron:              # iterates over keys by default
    print(key)

print()

for key, val in neuron.items():  # key-value pairs
    print(f"{key:20s} → {val}")
```

```python
# Nested dictionaries — common in neuroscience datasets
brain_atlas = {
    "hippocampus": {"lobe": "temporal",  "function": "memory",            "neurons": 40_000_000},
    "amygdala":    {"lobe": "temporal",  "function": "emotion",           "neurons": 13_000_000},
    "cerebellum":  {"lobe": "posterior", "function": "motor coordination","neurons": 69_000_000_000},
}

print(brain_atlas["hippocampus"]["function"])
print(brain_atlas["cerebellum"]["neurons"])
```

```python
# Building a dict from two lists using zip()
bands  = ["delta", "theta", "alpha", "beta", "gamma"]
ranges = ["0.5-4", "4-8",   "8-13",  "13-30", "30-100"]

eeg_bands = dict(zip(bands, ranges))
print(eeg_bands)
```

## 3.4 Sets

Sets hold **unique** elements with no guaranteed order. Fast for membership tests.

```python
# Sets eliminate duplicates automatically
recorded = ["cortex", "hippocampus", "cortex", "amygdala", "hippocampus"]
unique   = set(recorded)
print(unique)

# Note: sets are unordered — do not rely on order!
print(type(unique))
```

```python
# Membership test — sets are much faster than lists for this
print("hippocampus" in unique)
print("thalamus" in unique)
```

```python
# Set operations (like mathematical sets)
group_A = {"cortex", "hippocampus", "amygdala"}
group_B = {"hippocampus", "cerebellum", "amygdala"}

print("Union:       ", group_A | group_B)  # all elements
print("Intersection:", group_A & group_B)  # common elements
print("Difference:  ", group_A - group_B)  # in A but not B
```
