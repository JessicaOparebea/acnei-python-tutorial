We often want to store and work with *collections* of data — a list of spike times, a mapping from neuron ID to brain region, a set of unique conditions. Python provides four built-in data structures for this.

| Structure | Syntax | Ordered | Mutable | Unique items |
|-----------|--------|---------|---------|---------------|
| `list` | `[1, 2, 3]` | ✅ | ✅ | ❌ |
| `tuple` | `(1, 2, 3)` | ✅ | ❌ | ❌ |

## 3.1 Lists

A `list` is an ordered, mutable sequence. It is the most commonly used data structure in Python.

```python
# Creating lists
brain_regions = ["cortex", "hippocampus", "cerebellum", "amygdala", "thalamus"]
spike_times   = [0.1, 0.3, 0.5, 0.9, 1.2, 1.8]  # seconds
mixed         = ["cortex", 40, True, None]        # lists can hold any types
```

In Python, position 0=first, 1=second, -1=last.

```python
# Indexing — starts at 0!
print(brain_regions[0])    # first
```

```python
# Lists are MUTABLE — you can change them in place
brain_regions[0] = "prefrontal cortex"
print(brain_regions)
```

```python
# Useful list methods
regions = ["cortex", "hippocampus", "amygdala"]

regions.append("thalamus")       # add to end
print(regions)

regions.insert(1, "cerebellum")  # insert at position
print(regions)

regions.remove("amygdala")       # remove first occurrence of value
print(regions)

regions.pop(1)                   # remove and return last item
print(regions)
```

```python
# Checking for an element
print("hippocampus" in regions)
print("thalamus" in regions)
```

> **Important — Lists hold references, not copies:**

```python
a = [1, 2, 3]
b = a           # b points to the SAME list
b[0] = 99
print(a)        # [99, 2, 3] — a was also changed!

# To make a copy:
c = a.copy()
c[0] = 0
print(a)        # [99, 2, 3] — a is unchanged now
print(c)        # [0, 2, 3]
```

## 3.2 Tuples

Tuples are like lists but **immutable** — they cannot be modified after creation. Use them for data that should stay fixed.

```python
electrode_xyz = (34.5, -12.3, 8.0)   # x, y, z in mm — shouldn't change
print(electrode_xyz)
print(electrode_xyz[0])  # indexing works the same way
```

```python
# Tuple unpacking — very clean
x, y, z = electrode_xyz
print(f"Electrode: x={x}, y={y}, z={z}")

# Unpacking with *rest
first, *rest = (1, 2, 3, 4, 5)
print(first, rest)
```

## 3.3 Dictionaries

A dictionary maps **keys** to **values**. Keys must be unique and immutable (strings, numbers, tuples). Values can be anything.

```python
neuron = {}
neuron
```

```python
# Accessing values by key
print(neuron["type"])
```

```python
# Accessing a missing key raises a KeyError
try:
  print(neuron["axon"])
except KeyError as e:
    print("KeyError:",e)

# .get() is safer — returns None (or a default) if key is missing
print(neuron.get("axon_length", "unknown"))
```

```python
# Add / update keys
neuron["firing_rate_hz"] = 40
neuron["Vrest_mV"]       = -72   # update
print(neuron)
```

```python
# Iterating
for key in neuron:              # iterates over keys by default
    print(key)

print()

for key, val in neuron.items():  # key-value pairs
    print(f"{key:20s} → {val}")
```

## 3.4 Sets

Sets hold **unique** elements with no guaranteed order. Fast for membership tests.

```python
# Sets eliminate duplicates automatically
recorded = ["cortex", "hippocampus", "cortex", "amygdala", "hippocampus"]
unique   = set(recorded)
print(unique)

# Note: sets are unordered — do not rely on order!
print(type(unique))
```

```python
# Membership test — sets are much faster than lists for this
print("hippocampus" in unique)
print("thalamus" in unique)
```
